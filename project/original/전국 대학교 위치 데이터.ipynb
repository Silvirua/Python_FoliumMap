```python
import pandas as pd
import requests
from openpyxl import load_workbook
from openpyxl import Workbook
import re

filePath = r'/content/drive/MyDrive/python/Folium Map/학교 디지털 콘텐츠 공모전/2022_빅데이터/학교개황.xls'
df_from_excel = pd.read_excel(filePath)

url = 'http://api.vworld.kr/req/address?'
params = 'service=address&request=getcoord&version=2.0&crs=epsg:4326&refine=true&simple=false&format=json&type='
road_type = 'ROAD'     # 도로명 주소
road_type2 = 'RARCEL'  # 지번 주소
address = '&address='
keys = '&key='
primary_key = "ED1762FF-6965-30D3-8180-51754A5E382D"

def request_geo(road):
    page = requests.get(url+params+road_type+address+road+keys+primary_key)
    json_data = page.json()
    if json_data['response']['status'] == 'OK':
        x = json_data['response']['result']['point']['x']
        y = json_data['response']['result']['point']['y']
        return x, y
    else:
        x = 0
        y = 0
        return x, y

try:
    wb = load_workbook(r'/content/drive/MyDrive/python/Folium Map/학교 디지털 콘텐츠 공모전/2022_빅데이터/학교주소좌표.xlsx', data_only=True)
    sheet = wb.active
except:
    wb = Workbook()
    sheet = wb.active

university_list = df_from_excel['학교명'].to_list()
address_list = df_from_excel['주소'].to_list()

for num, value in enumerate(address_list):
    addr = re.sub(r'/([^)]*/)', '', value)
    print(addr)
    x, y = request_geo(addr)
    sheet.append([university_list[num], addr, x, y])

wb.save(r'/content/drive/MyDrive/python/Folium Map/학교 디지털 콘텐츠 공모전/2022_빅데이터/학교주소좌표.xlsx')
```
